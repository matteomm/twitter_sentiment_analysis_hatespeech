```python
# create function to strip emojis from the stream
def removeEmoj(text):
    if text:
        return text.encode('ascii', 'ignore').decode('ascii')
    else:
        return None
```

```python
# create a function to store data in a MYSQL database
def store_data(id_str , created_at, text, processed):
    db=mysql.connector.connect(host="localhost", user="root", passwd="password", db="twitter_2", charset="utf8")
    cursor = db.cursor()
    insert_query = "INSERT INTO twitter_2 (id_str, created_at, text, processed) VALUES (%s, %s, %s, %s)"
    cursor.execute(insert_query, (id_str , created_at, text,processed))
    db.commit()
    cursor.close()
    db.close()
    return
```
